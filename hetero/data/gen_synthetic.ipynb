# Generate Synthetic Data from Real Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
RAW_DATA = pd.read_pickle("rl_samples.pkl")
```

```python
print(RAW_DATA.shape)
print(RAW_DATA.columns)
```

# Generate Pseudo data

```python
mu = 0      # mean
sigma = 1   # standard deviation

PSEUDO_RAW_DATA = pd.DataFrame(sigma * np.random.randn(*RAW_DATA.shape) + mu, index=RAW_DATA.index, columns=RAW_DATA.columns)
```

```python
PSEUDO_RAW_DATA_PART = PSEUDO_RAW_DATA.iloc[:500, :]
```

```python
with open("pseudo_rl_samples.pkl", "wb") as file:
    pickle.dump(PSEUDO_RAW_DATA, file)
```
